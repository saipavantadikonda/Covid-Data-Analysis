Importing pandas library to read and merge csv files 

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

Reading cases csv file and displaying the cases dataframe

```python
covid_cases = pd.read_csv("../../../data/Stage_1/covid_confirmed_usafacts.csv")
covid_cases
```

```python
covid_cases = covid_cases[covid_cases['County Name'] != 'Statewide Unallocated']
covid_cases
```

```python
covid_cases['County Name'] = covid_cases['County Name'].str.strip()
covid_cases
```

Reading deaths csv file and displaying the deaths dataframe

```python
covid_deaths = pd.read_csv("../../../data/Stage_1/covid_deaths_usafacts.csv")
covid_deaths 
```

```python
covid_deaths = covid_deaths[covid_deaths['County Name'] != 'Statewide Unallocated']
covid_deaths
```

```python
covid_deaths['County Name'] = covid_deaths['County Name'].str.strip()
covid_deaths
```

Reading population csv file and displaying the population dataframe

```python
covid_county_population = pd.read_csv("../../../data/Stage_1/covid_county_population_usafacts.csv")
covid_county_population
```

```python
covid_county_population = covid_county_population[covid_county_population['County Name'] != 'Statewide Unallocated']
covid_county_population
```

```python
covid_county_population['County Name'] = covid_county_population['County Name'].str.strip()
covid_county_population
```

```python
covid_county_population = covid_county_population.drop('County Name',axis = 1)
covid_county_population
```

Data types of covid_deaths dataframe

```python
covid_deaths.dtypes
```

Data types of county population dataframe

```python
covid_county_population.dtypes
```

Merging covid_cases with covid_deaths dataframes as cases_deaths dataframe on common columns countyFIPS, County Name, State, StateFIPS.

```python
cases_deaths = pd.merge(covid_cases, covid_deaths, on = ["countyFIPS",'County Name', "State", "StateFIPS"])
cases_deaths
```

Merging cases_deaths with covid_county_population dataframe as super_covid19 dataframe on columns countyFIPS, State 

```python
super_covid19 = pd.merge(cases_deaths, covid_county_population, on = ["countyFIPS","State"])
super_covid19
```

The super_covid19 dataframe is exported as super_covid19.csv file

```python
super_covid19.to_csv("../../../data/Stage_1/super_covid19.csv", index = False)
```

```python
covid_cases_long = pd.melt(covid_cases, id_vars = ['countyFIPS','County Name','State','StateFIPS'],var_name ='Dates', value_name = 'cases')
covid_cases_long
```

```python
covid_deaths_long = pd.melt(covid_deaths, id_vars = ['countyFIPS','County Name','State','StateFIPS'],var_name ='Dates', value_name = 'deaths')
covid_deaths_long
```
