Reading super covid csv file and displaying the dataframe

Reading ecnomic dataset csv file and displaying the dataframe

```python
economic_df = pd.read_csv("/Users/anshu/Desktop/Courses/Data Science/Project Stage - I/DP03/ACSDP5Y2021_DP03-RawData.csv", low_memory= False)
economic_df
```

Removing columns with NaN values

```python
# find and drop the column names that contain the 'MA' text
columns_to_drop = [col for col in economic_df.columns if "MA" in col]
economic_df = economic_df.drop(columns_to_drop, axis=1)

# find and drop the column names that end with the letter 'M'
columns_to_drop2 = [col for col in economic_df.columns if col[-1] == 'M']
economic_df = economic_df.drop(columns_to_drop2, axis=1)

# find and drop the column names that contain the 'EA' text
columns_to_drop3 = [col for col in economic_df.columns if "EA" in col]
economic_df = economic_df.drop(columns_to_drop3, axis=1)

economic_df
```

Removing row 0
Extracting countyFIPS from GEO_ID by splitting at S
Splitting county and state name by splitting from ,

```python
#removing 0th row
economic_df = economic_df.drop(economic_df.index[0])

#Extracting countyFIPS from GEO_ID by splitting at S
economic_df[['col1', 'countyFIPS']] = economic_df['GEO_ID'].str.split("S", expand=True)
economic_df = economic_df.drop('GEO_ID', axis=1)
economic_df = economic_df.drop('col1', axis=1)

#Splitting county and state name by splitting from ,
economic_df[['County Name', 'State']] = economic_df['NAME'].str.split(",", expand=True)
economic_df = economic_df.drop('NAME', axis=1)

#removing any extra spaces
for col in economic_df.columns:
    if economic_df[col].dtype == object:
        economic_df[col] = economic_df[col].str.strip()

economic_df
```

Selecting the columns which will be useful for covid analysis

```python
economic_df = economic_df[['countyFIPS','County Name','State', 'DP03_0001E', 'DP03_0010E', 'DP03_0021E', 'DP03_0024E', 'DP03_0029E', 'DP03_0042E', 'DP03_0048E', 'DP03_0050E', 
                         'DP03_0052E', 'DP03_0061E', 'DP03_0086E', 'DP03_0097E','DP03_0099E', 'DP03_0106E', 'DP03_0108E', 'DP03_0134E', 'DP03_0135E']]

economic_df
```

Renaming the columns name and changing datatype of countyFIPS to int64 for merging with super covid dataset

```python
economic_df.rename(columns={'State' : 'State_USA',
                            'DP03_0001E' : ' Employment Status-Employed(Above 16 years)', 
                            'DP03_0010E' : 'Employment Status-Employed(Female)', 
                            'DP03_0021E' : ' Commuting to work with Public transport', 
                            'DP03_0024E' : 'Working from home', 
                            'DP03_0029E' : ' Occupation-Sales and office occupations',
                            'DP03_0042E' : ' Industry-Educational services, and health care and social assistance',
                            'DP03_0048E' : 'Class of worker-Government workers', 
                            'DP03_0050E' : 'Unpaid family workers', 
                            'DP03_0052E' : 'Income-Less than $10,000', 
                            'DP03_0061E' : 'Income-$200,000 or more', 
                            'DP03_0086E' : 'Median family income', 
                            'DP03_0097E' : 'With private health insurance',
                            'DP03_0099E' : 'No health insurance coverage', 
                            'DP03_0106E' : 'With private health insurance(Employed noninstitutionalized population 19 to 64 years)', 
                            'DP03_0108E' : 'No health insurance coverage(Employed noninstitutionalized population 19 to 64 years)', 
                            'DP03_0134E' : '18 to 64 years (People whose income in past 12 years is below poverty level)', 
                            'DP03_0135E' : '65 years and over (People whose income in past 12 years is below poverty level)'},
          inplace=True)
economic_df['countyFIPS'] = economic_df['countyFIPS'].astype('int64')
economic_df
```
