```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid_cases_df = pd.read_csv("/Users/anshu/Desktop/Courses/Data Science/Project Stage - I/covid_confirmed_usafacts.csv", low_memory= False)
covid_cases_df
```

```python

covid_cases_df = covid_cases_df[covid_cases_df['State'] == 'AK']

covid_cases_df = covid_cases_df[['2023-01-10', '2023-01-11',
                                 '2023-01-12','2023-01-13', '2023-01-14', '2023-01-15', 
                                 '2023-01-16']]

covid_cases_df
```

Calculating sum for each date

```python
# Calculate the sum of each column
sums = covid_cases_df.sum().tolist()
sums
```

```python
# Plot to display the column sums
plt.plot(covid_cases_df.columns, sums)
plt.xlabel('Dates')
plt.ylabel('Total Cases')
plt.title('Cases in last week')
plt.show()
```

Covid cases for Alaska is increasing continuously in first few days of week, from 01/14 to 01/16 cases are stable.
