### Calculate COVID-19 data trends for last week of the data. Are the cases increasing, decreasing, or stable? Each student chooses a state to analyze:

Importing the pandas and matplotlib.pyplot libraries, which are widely used for data manipulation and visualization in Python.
Pandas library for reading csv file and performing analysis on that data.matplotlib for plotting the cases last week trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Reading covid cases csv file and displaying the cases data

```python
covid_cases = pd.read_csv("../../../../data/Stage_1/covid_confirmed_usafacts.csv")
covid_cases
```

Just describing the data type of each columns, it's important to know the data types of each column so we can properly analyze and manipulate the data.

```python
covid_cases.dtypes
```

Here we are just getting result in a new dataframe that contains only the rows where the "County Name" column is not equal to "Statewide Unallocated". The new dataframe is then assigned back to the "covid_cases" variable.

```python
covid_cases = covid_cases[covid_cases['County Name'] != 'Statewide Unallocated']
```

Creating a new dataframe that only contains the COVID-19 cases specifically in the state of North Carolina, for further analysis  of last week covid cases trend.

```python
covid_cases_NC = covid_cases[covid_cases['State'] == 'NC']
```

```python
covid_cases_NC
```

Creating a DataFrame that contains the last week data for the state of North Carolina.

```python
last_covid_trends=covid_cases_NC[covid_cases_NC.columns[-7:]]
last_covid_trends
```

Calulating last week data by using a "sum()" method that computes the sum of all values in a DataFrame or a series.The "axis" parameter specifies the axis along which the sum is calculated, with 0 representing the rows and 1 representing the columns.
So, in this case, the sum of the last week data is calculated for each row of the "last_covid_trends" DataFrame and stored as a Series in the "total" variable.

```python
total = last_covid_trends.sum(axis = 0)
```

```python
total
```

We are using  reset_index() method to reset the index of the covid_trends DataFrame to a default range index, so that the index of the total series is converted into a column in the new DataFrame (covid_trends).

```python
covid_trends = pd.DataFrame(data = total).reset_index() 
```

We are renaming the columns of the data frame "covid_trends". The original name of the first column (index column) is changed to "Date" and the original name of the second column (0th column) is changed to "Cases". The "inplace" argument is set to "True", meaning the changes are made directly to the original data frame, rather than creating a new data frame with the changes.

```python
covid_trends.rename(columns={'index':'Date',0:"Cases"},inplace = True)
```

```python
covid_trends  # The DataFrame represents a table of data
```

We are plotting a graph using the plot method , with the x-axis being the "Date" column and the y-axis being the "Cases" column. The figure size is set to 10 by 5, and the title of the graph is set to "Covid Cases Trends for last week in NC State". Finally, the x-axis and y-axis labels are set to "Date" and "Cases", respectively. The plt.show() function is used to display the graph.

```python
ax=covid_trends.plot(x="Date",y="Cases",figsize=(10,5),title='Covid Cases Trends for last week in NC State')
ax.set(xlabel="Date",ylabel="Cases")
plt.show()
```

* From the above graph we can analyze how the cases are Stable from January 10th to January 11 2023.
* We can observe a sudden increase in the cases on January 11th to 12th , Almost 22624 cases got registered NC state During that period of time.
* Later on the cases became stable till January 16th.
