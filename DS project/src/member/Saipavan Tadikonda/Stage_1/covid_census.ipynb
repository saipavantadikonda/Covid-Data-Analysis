Importing pandas library 

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

Reading census demographic csv file and displaying the dataframe

```python
census_demographic = pd.read_csv("../../../../data/Stage_1/ACSDP5Y2021.DP05-Data.csv", low_memory= False)
census_demographic
```

```python
census_demographic.drop(index= 0, inplace = True)   # dropping the zeroth row as it contains only titles
census_demographic
```

Removing unwanted columns from the dataframe as it doesn't contribute to any covid analysis

```python
census_demographic.drop(census_demographic.columns[census_demographic.columns.str.contains('EA')], axis= 1, inplace = True)
census_demographic.drop(census_demographic.columns[census_demographic.columns.str.endswith('M')], axis=1, inplace = True)
census_demographic.drop(census_demographic.columns[census_demographic.columns.str.contains('MA')], axis = 1, inplace= True)
census_demographic.drop(census_demographic.columns[census_demographic.columns.str.contains('PE')], axis = 1, inplace= True)
census_demographic.drop(census_demographic.columns[census_demographic.columns.str.contains('Unnamed')], axis = 1, inplace= True)
census_demographic
```

Data types of census demographic dataframe

```python
census_demographic.dtypes
```

In order to merge with super covid19 dataframe we need to find the common columns in both dataframes. If we observe GEO_ID column of census dataframe we can see that it contains unique five digit countyFIPS in it. 

```python
#slicing the GEO_ID from 9th inorder to obtain countyFIPS
census_demographic['GEO_ID']= census_demographic['GEO_ID'].str.slice(9) 
#changing the datatype of GEO_ID to int64 as it should match with the super_covid19 column
census_demographic['GEO_ID']= census_demographic['GEO_ID'].astype('int64') 
#Renaming the GEO_ID to countyFIPS as it should match with the countyFIPS column in the super_covid19 dataframe
census_demographic.rename(columns = {'GEO_ID':'countyFIPS'}, inplace = True) 
#Removing the extra rows as it doesnt match with the covid19 dataset
census_demographic = census_demographic[census_demographic['countyFIPS'] < 72001] 
census_demographic
```

If we observe the NAME column in census_demographic dataframe it contains both county name and state. Splitting them at comma using the split function. 

```python
census_demographic[['County Name','State']]= census_demographic['NAME'].str.split(',',1, expand= True)
#Removing the unwanted spaces to the left and right of the State
census_demographic['State'] = census_demographic['State'].str.strip() 
census_demographic
```

Reading the generated super_covid19 dataset in the team task inorder to merge with the enrichment dataset

```python
super_covid19 = pd.read_csv('../../../../data/Stage_1/super_covid19.csv')
super_covid19
```

Renaming the County Name_x to County Name as the name should match for merging and dropping the County Name_y from the dataframe

```python
super_covid19.rename(columns={'County Name_x':'County Name'},inplace = True)
super_covid19.drop('County Name_y',axis=1,inplace = True)
super_covid19
```

Removing the unwanted spaces from the County Name if there exists any

```python
super_covid19['County Name'] = super_covid19['County Name'].str.strip()
super_covid19
```

Removing the rows whose county name is statewide unallocated

```python
super_covid19 = super_covid19[super_covid19['County Name']!= 'Statewide Unallocated']
super_covid19
```

Removing unwanted spaces in county name column of census_demographic dataframe if there exists any

```python
census_demographic_temp['County Name'] = census_demographic_temp['County Name'].str.strip()
census_demographic['County Name'] = census_demographic['County Name'].str.strip()
```

Merging the super_covid19 dataframe and census_demographic_temp dataframe on countyFIPS, County Name, State columns as they are common in both dataframes

```python
covid_census_temp = pd.merge(super_covid19,census_demographic_temp,on = ['countyFIPS','County Name','State'])
covid_census_temp
```

Merging the super_covid19 dataframe and census_demographic dataframe on countyFIPS, County Name and State columns as they are common in both

```python
covid_census = pd.merge(super_covid19,census_demographic,on = ['countyFIPS','County Name','State'])
covid_census
```

Exporting the merged covid_census dataframe in the csv format 

```python
covid_census.to_csv('../../../../data/Stage_1/covid_census.csv',index= False)
```
