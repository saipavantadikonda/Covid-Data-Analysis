### Calculate COVID-19 data trends for last week of the data. Are the cases increasing, decreasing, or stable? Each student chooses a state to analyze.

Importing pandas library for reading csv file and performing analysis on that data.
Importing matplotlib for plotting the cases last week trends

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Reading covid cases csv file and displaying the cases data

```python
covid_cases = pd.read_csv("../../../../data/Stage_1/covid_confirmed_usafacts.csv")
covid_cases
```

Removing records whose County Name is Statewide Unallocated as it contains all zeros

```python
covid_cases_final = covid_cases[covid_cases['County Name']!= 'Statewide Unallocated']
covid_cases_final
```

Selected California State(CA) for analyzing last week covid cases trend

```python
covid_cases_ca = covid_cases_final[covid_cases_final['State']=='CA']
covid_cases_ca
```

Separating last week data of California state into another dataframe using iloc() function

```python
covid_cases_ca_lw = covid_cases_ca.iloc[:,-7:]
covid_cases_ca_lw
```

Calculating sum of cases on each day of last week using sum

```python
covid_cases_ca_lw_sum = covid_cases_ca_lw.sum(axis=0)
covid_cases_ca_lw_sum
```

```python
cases_trends_ca = pd.DataFrame(covid_cases_ca_lw_sum) # Creating new dataframe
cases_trends_ca
```

```python
cases_trends_ca.reset_index(inplace = True)  # Resetting index
cases_trends_ca
```

```python
cases_trends_ca.rename(columns = {'index':'Dates', 0:'Cases'}, inplace = True) #Renaming columns of the cases_trends_ca dataframe
cases_trends_ca
```

Plotting the last week covid 19 cases data of California State using matplotlib plot function

```python
plt.plot(cases_trends_ca['Dates'],cases_trends_ca['Cases'])
plt.xlabel('Dates')
plt.ylabel('Cases')
plt.title('Covid-19 last week trends for California State')
plt.show()
```

- From the above plot we can observe that the cases in California State(CA) increased day by day in the last week i.e., from 10th Jan to 16th Jan 2023.

- 15171 new cases got registered in CA State in the last week
