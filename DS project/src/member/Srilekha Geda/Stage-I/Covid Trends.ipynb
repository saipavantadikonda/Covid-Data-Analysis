Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Reading confirmed caes dataset

```python
c=pd.read_csv(r"..\..\..\..\data\Stage_1\covid_confirmed_usafacts.csv")
c
```

New Jersey state details 

```python
NJpcd=c.loc[c["State"]=="NJ"]
NJpcd
```

```python
s=[0 for i in range(7)]
for i in range(7):
    seven=c.columns[-(i+1)]
    s[i]=seven
    
s_inv=s[::-1]
s_inv
```

Sum of the last week covid cases

```python
s_out=[0 for i in range(len(s))]
for i in range(len(s)):
    sum=0
    for j in range(NJpcd.shape[0]):
#         print(s[i])
#         print(NJpcd[s[i]].iloc[j])
        sum=sum+NJpcd[s_inv[i]].iloc[j]
#         print(sum)
    s_out[i]=sum
print(s_out)
```

```python
fig, ax= plt.subplots(1)
ax.plot(s_inv,s_out,color='r')
fig.autofmt_xdate()
plt.show()
```
