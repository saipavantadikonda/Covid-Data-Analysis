Importing Libraries

```python
import numpy as np
import pandas as pd
```

Reading the Enrichment Data

```python
elc=pd.read_csv(r"..\..\..\..\data\Stage_1\president_county_candidate.csv")
elc
```

```python
elc.dtypes
```

Reading the states codes files and convert to dictionary

```python
pa=pd.read_excel(r"..\..\..\..\data\Stage_1\postal_abbr.xlsx")
pad=pa.set_index('state').to_dict()['state_abbr']
pad
```

Mapping the elections data set with sate codes(abbrevations)

```python
elc["state_ab"]=elc["state"].map(pad)
elc
```

Reading the SUPER COVID data set

```python
sc=pd.read_csv(r"..\..\..\..\data\Stage_1\super_covid19.csv")
sc
```

```python
sc['County Name_x'] = sc['County Name_x'].str.strip()
```

Merging the elctions dataset to the super covid dataset using county and state

```python
pc=pd.merge(elc,sc,left_on=["county","state_ab"],right_on=['County Name_x',"State"])
pc
```

```python
pc.isnull().values.any()
```

Merging the elctions dataset to the super covid dataset using county and state with left join

```python
pc1=pd.merge(elc,sc,left_on=["county","state_ab"],right_on=['County Name_x',"State"],how='left')
pc1
```
